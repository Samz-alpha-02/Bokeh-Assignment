```python
#Q2
In Bokeh, glyphs are the visual elements that represent the data in a plot. They can be used to display various
shapes, markers, lines, and other graphical elements to represent the data points or patterns in a plot. Glyphs
allow you to customize the appearance and style of the data representation in the plot.
```

```python
from bokeh.plotting import figure, show

# Create a new figure
p = figure(title='My Bokeh Plot', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add circle and line glyphs to the plot
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=10, color='blue')
p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2, color='red')

# Display the plot
show(p)
```

```python
In this example, we import the figure() and show() functions from bokeh.plotting to create and display
the plot, respectively.

We create a new figure using figure() and provide a title, x-axis label, and y-axis label.

Next, we add two glyphs to the plot. First, we add a circle glyph using the circle() function. We pass
the x-coordinates, y-coordinates, size, and color for the circle markers.

Then, we add a line glyph using the line() function. We pass the x-coordinates, y-coordinates, line width,
and color for the line.

Finally, we use show() to display the plot.

In this example, we added both circle and line glyphs to the plot. The circle glyphs represent individual
data points, while the line glyph represents a line connecting the data points. You can customize various
attributes of the glyphs, such as size, color, line width, and more, to suit your visualization needs.
```

```python
#Q3
In Bokeh, you can customize the appearance of a plot, including the axes, title, and legend, using various
options and properties provided by the figure object. Here are some common ways to customize the appearance
of a Bokeh plot:

1.Axes customization:

a.Set axis labels: Use the x_axis_label and y_axis_label properties of the figure object to set the labels for
                   the x-axis and y-axis, respectively.
b.Set axis ranges: Use the x_range and y_range properties of the figure object to set the range of values displayed
                   on the x-axis and y-axis, respectively.
c.Customize tick labels and formatting: You can customize tick labels, their orientation, and formatting using the
                                        xaxis and yaxis properties of the figure object. For example,
                                        figure.xaxis.major_label_orientation can be used to set the orientation of
                                        x-axis tick labels.
d.Set axis line properties: You can customize the appearance of axis lines, ticks, and tick lines using properties
                            like xaxis.axis_line_width, yaxis.major_tick_line_color, etc.

2.Title customization:

a.Set plot title: Use the title property of the figure object to set the title of the plot.
b.Customize title appearance: You can customize the appearance of the plot title, such as font size, color,
                              and alignment, using properties like title.text_font_size, title.text_color, and title.align.

3.Legend customization:

a.Add a legend: To add a legend to the plot, use the legend property of the figure object and specify the location
                using values like "top_left", "bottom_right", etc.
b.Customize legend appearance: You can customize the appearance of the legend, such as font size, background color,
                               and border, using properties like legend.label_text_font_size, legend.background_fill_color,
                               and legend.border_line_color.
```

```python
from bokeh.plotting import figure
from bokeh.io import show

# Create a new figure
p = figure(title='Customized Bokeh Plot', x_axis_label='X-axis', y_axis_label='Y-axis',
           width=500, height=400)

# Customize axes
p.xaxis.axis_label_text_font_size = '14pt'
p.yaxis.axis_label_text_color = 'blue'
p.xaxis.axis_line_width = 2
p.yaxis.major_tick_line_color = 'green'

# Customize title
p.title.text_font_size = '16pt'
p.title.text_color = 'red'
p.title.align = 'center'

# Customize legend
p.legend.location = 'top_right'
p.legend.label_text_font_size = '12pt'
p.legend.background_fill_color = 'lightyellow'
p.legend.border_line_color = 'orange'

# Add glyphs to the plot (example circles)
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=10, color='blue', legend_label='Data')

# Display the plot
show(p)
```

```python
#Q4
A Bokeh server is a Python library that allows you to create and deploy interactive data applications with Bokeh.
It enables you to build web-based dashboards, data visualization tools, and applications that can update and respond
to user interactions in real time.

Using the Bokeh server, you can create interactive plots that can be updated dynamically as users interact with
the plot or as new data becomes available. The server runs a Python script that handles the plot creation, data
processing, and event handling. It communicates with a web browser to render and update the plot in real time.

To use the Bokeh server and create interactive plots:

1.Define the layout and plot elements: Create the layout for your application using the bokeh.layouts module.
                                       This can include multiple plots, widgets, buttons, sliders, and other
                                       UI components.

2.Set up callbacks: Define callback functions that respond to user interactions or changes in data. These
                    callbacks can update the plot, modify data, or trigger other actions.

3.Run the Bokeh server: Start the Bokeh server by running the Python script. The server runs locally and
                        serves the application to a web browser.

4.View and interact with the plot: Open a web browser and navigate to the server address (usually http://localhost:5006).
                                   You can view and interact with the plot in the web browser. Any interactions or
                                   data updates trigger the callbacks, allowing the plot to be updated dynamically.
```

```python
from bokeh.plotting import curdoc, figure
from bokeh.models import Slider

# Create a figure
p = figure()

# Create a slider widget
slider = Slider(start=0, end=10, step=0.1, value=5, title="Slider")

# Define a callback function
def update_plot(attr, old, new):
    p.circle([1, 2, 3, 4, 5], [new, new+1, new+2, new+3, new+4], size=10)

# Register the callback with the slider widget
slider.on_change('value', update_plot)

# Add the slider and plot to the document
curdoc().add_root(slider)
curdoc().add_root(p)
```
